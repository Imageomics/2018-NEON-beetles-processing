```python
import cv2
import numpy as np
import os
```

```python

# Define the directory path
directory_path = 'D:/University/RA/Beetle_Palooza/Data/Full_Data/2018-NEON-beetles/group_images/'

# List to store image paths
image_paths = []

# Loop through the directory to get all files
for filename in os.listdir(directory_path):
    image_paths.append(directory_path + filename)

len(image_paths)
```

```python
# Load the image

# input_image_directory = "D:/University/RA/Beetle_Palooza/Data/Full_Data/2018-NEON-beetles/group_images/"
output_directory = "D:/University/RA/Beetle_Palooza/Data/separate_images/"
# image = cv2.imread('D:/University/RA/Beetle_Palooza/Data/BeetleImage-A00000033619.jpg')

image_index = 1

image = cv2.imread(image_paths[image_index])
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply a threshold to get a binary image
_, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)

# Find contours in the image
contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Parameters for blue color detection
blue_threshold = 200  # Blue intensity threshold
blue_pixel_percentage_threshold = 5  # Ignore images if more than 5% of pixels are blue

min_file_size = 30 * 1024  # 20KB
max_file_size = 700 * 1024  # 20KB

# Loop over contours and save each beetle image
for i, contour in enumerate(contours):
    # Get bounding box for each contour
    x, y, w, h = cv2.boundingRect(contour)
    
    # Extract beetle from the image
    beetle = image[y:y+h, x:x+w]
    
    # Save the extracted beetle image
#     file_path = f'D:/University/RA/Beetle_Palooza/Data/separate_images/beetle_{i}.png'
    image_name_final = image_paths[0].split("/")[-1].split(".")[image_index]
    file_path_name = output_directory + image_name_final
    file_path = output_directory + image_name_final + "/beetle_" + str(i) + ".png"
    
    # Create the directory if it does not exist
    if not os.path.exists(file_path_name):
        os.makedirs(file_path_name)
    
    cv2.imwrite(file_path, beetle)

    file_size = os.path.getsize(file_path)
    # Check the file size
    if file_size < min_file_size or file_size > max_file_size:  # 10KB = 10 * 1024 bytes
        os.remove(file_path)  # Delete the image if it's smaller than 10KB
        continue

    # Check for blue color in the image
    blue_channel = beetle[:, :, 0]  # Blue channel of the image
    total_pixels = beetle.size // 3  # Total number of pixels in the image
    blue_pixels = cv2.inRange(blue_channel, blue_threshold, 255)
    blue_pixel_count = cv2.countNonZero(blue_pixels)
    
    # Calculate the percentage of blue pixels
    blue_pixel_percentage = (blue_pixel_count / total_pixels) * 100
    
    # If the percentage of blue pixels exceeds the threshold, delete the image
    if blue_pixel_percentage > blue_pixel_percentage_threshold:
        os.remove(file_path)  # Delete the image if it contains too much blue color

        
print(f"Extracted {len(contours)} beetle images.")
```
